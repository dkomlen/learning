```python
from math import floor, ceil

import numpy as np
from sympy import *
import matplotlib.pyplot as mpl

%matplotlib inline
```

1.2-2 Suppose we are comparing implementations of insertion sort and merge sort on the same machine. For inputs of size `n`, insertion sort runs in `8n^2` steps, while merge sort runs in `64n lgn` steps. For which values of `n` does insertion sort beat merge sort?

```python
n = symbols('n')

eq = log(n,2)-n/8
sol = [N(x) for x in solve(eq, n)]
sol = [ceil(sol[0]), floor(sol[1])]
print('solution: [{},{}]'.format(*sol))
```

```python
fig = mpl.figure()
ax = fig.add_subplot(111)

xs = np.linspace(0, 50, 100)
fx = lambdify(n, eq, modules=['numpy'])

ax.plot(xs, fx(xs))
ax.scatter(sol, [0,0], color='red', marker='x')

ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
```

1.2-3 What is the smallest value of `n` such that an algorithm whose running time is `100n^2` runs faster than an algorithm whoose running time is `2^n` on the same machine?

```python
eq = 100*n**2-2**n
sol = [N(x) for x in solve(eq, n)]
sol = [floor(sol[0]), ceil(sol[1])]
print('solution: [{},{}]'.format(*sol))
```

```python
fx = lambdify(n, eq, modules=['numpy'])
fig = mpl.figure()
ax = fig.add_subplot(111)
xs = np.linspace(0, 20, 100)
ax.plot(xs, fx(xs))
ax.scatter(sol, [0,0], color='red', marker='x')

ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
mpl.ylim(bottom=-1000);
```
